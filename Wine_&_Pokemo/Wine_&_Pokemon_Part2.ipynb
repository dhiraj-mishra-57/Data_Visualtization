## Subplotting

`Subplotting` is a technique for creating multiple plots that live side-by-side in one overall figure. We can use the subplots method to create a figure with multiple subplots. subplots takes two arguments. The `first` one controls the `number of rows`, the `second` one the `number of columns`

```python
fig,ax = plt.subplots(2,1)
```

```python
ax
```

```python
fig,axis = plt.subplots(1,2,figsize=(12,10))
wine['province'].value_counts().head(10).plot.bar(fontsize=10,ax=axis[0]) #Top 10 Wine producing provience
new_wine['province'].value_counts().head(10).plot.bar(ax=axis[1]) #Wine Produced in US only
```

```python
pokemon.head()
```

```python
pokemon.loc[:,['speed','weight_kg']]
pokemon.plot.scatter(x='speed',y='weight_kg')
```

```python
#Let's now add new fetaures in the graph
fig,axis = plt.subplots(1,2,figsize=(15,10))
axis[0].set_title('Pokemon')
axis[1].set_title('Pokemon')
pokemon.plot.scatter(x='speed',y='weight_kg',ax=axis[0])
pokemon.loc[:10,['name','speed']].plot.bar(x='name',ax=axis[1])
```

```python
#Change explode value and observe the changes in the graph
explode=[0.1,0.01,0.1,0.01,0.1,0.01,0.1,0.01,0.1,0.01]
fig,axis = plt.subplots(1,2,figsize=(15,10))
axis[0].set_title('Pokemon')
axis[1].set_title('Pokemon')
pokemon.plot.scatter(x='speed',y='weight_kg',ax=axis[0])
pokemon.loc[:9,['name','speed']].plot.pie(y='speed',ax=axis[1],explode=explode)
plt.legend(loc=2) #Location of legent at the top of graph
```
